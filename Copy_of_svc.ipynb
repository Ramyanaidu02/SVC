```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
```

```python
X,y = make_blobs(n_samples=100,n_features=2,centers=2,cluster_std=1.05,random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=100,cmap='autumn')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
clf = SVC(kernel = 'poly')
```

```python
clf.fit(X_train,y_train)
```

```python
plot_decision_regions(X_train,y_train,clf=clf,legend=2)
support_vectors = clf.support_vectors_
plt.scatter(support_vectors[:,0],support_vectors[:,1],s=100,linewidth=1,facecolors='none',edgecolors='k')
plt.show()
```

```python
print(support_vectors)
```
